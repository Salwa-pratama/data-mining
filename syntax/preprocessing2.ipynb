## 2. Encoding 

### Data Kategorikal (String)
- Property Area: Urban, Rural, Semiurban  
- Loan Status: Y, N  
- Self Employed: No, Yes  
- Education: Graduate, Not Graduate  
- Married: No, Yes  
- Gender: Male, Female
- Dependents : 0 , 1 , 2 , 3+

Data set asli

```python
df_encoded = df.copy()

# Binary encoding
df_encoded['Gender'] = df_encoded['Gender'].map({'Male': 1, 'Female': 0})
df_encoded['Married'] = df_encoded['Married'].map({'Yes': 1, 'No': 0})
df_encoded['Education'] = df_encoded['Education'].map({'Graduate': 1, 'Not Graduate': 0})
df_encoded['Self_Employed'] = df_encoded['Self_Employed'].map({'Yes': 1, 'No': 0})
df_encoded['Loan_Status'] = df_encoded['Loan_Status'].map({'Y': 1, 'N': 0})
df_encoded['Property_Area'] = df_encoded['Property_Area'].map({'Urban': 1, 'Rural': 2, "Semiurban" : 3})

```

dataset yang outliersnya direplace

```python
df_encoded_outliers_replace = df_outliers_replace.copy()

# Binary encoding
df_encoded_outliers_replace['Gender'] = df_encoded_outliers_replace['Gender'].map({'Male': 1, 'Female': 0})
df_encoded_outliers_replace['Married'] = df_encoded_outliers_replace['Married'].map({'Yes': 1, 'No': 0})
df_encoded_outliers_replace['Education'] = df_encoded_outliers_replace['Education'].map({'Graduate': 1, 'Not Graduate': 0})
df_encoded_outliers_replace['Self_Employed'] = df_encoded_outliers_replace['Self_Employed'].map({'Yes': 1, 'No': 0})
df_encoded_outliers_replace['Loan_Status'] = df_encoded_outliers_replace['Loan_Status'].map({'Y': 1, 'N': 0})
df_encoded_outliers_replace['Property_Area'] = df_encoded_outliers_replace['Property_Area'].map({'Urban': 1, 'Rural': 2, "Semiurban" : 3})
```

## Decimal Scaling

```python
def decimal_scaling(df, cols):
    df_scaled = df.copy()

    for col in cols:
        max_val = df_scaled[col].abs().max()
        j = len(str(int(max_val)))
        df_scaled[col] = df_scaled[col] / (10 ** j)

    return df_scaled
```

Apply

```python
df_decimal = decimal_scaling(df_encoded, num_cols)
df_decimal_drop = decimal_scaling(df_encoded_clean_outliers, num_cols)
df_decimal_replace = decimal_scaling(df_encoded_outliers_replace, num_cols)
```
